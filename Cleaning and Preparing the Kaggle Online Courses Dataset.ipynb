### Détection des valeurs manquantes:


```python
data.isnull().sum()
```

```python
num_cols = data.select_dtypes(include=['int64', 'float64']).columns
cat_cols = data.select_dtypes(include=['object', 'bool']).columns
print("VARIABLES QUANTITATIVES:", num_cols.tolist())
print("VARIABLES QUALITATIVES:", cat_cols.tolist())
```

### Nettoyage des valeurs manquantes:

### a) Supprimer la colonne "Unique Projects" (100% vide):


```python
data = data.drop('Unique Projects', axis=1)
print("la colonne 'Unique Projects' est supprimée ")
# Mettre à jour num_cols après suppression
num_cols = data.select_dtypes(include=['int64', 'float64']).columns
print("Variables quantitatives après suppression:", num_cols.tolist())
```

### b)Nettoyer les variables quantitatives avec médiane:

```python
data[num_cols] = data[num_cols].apply(lambda x: x.fillna(x.median()))
print("✓ Variables quantitatives nettoyées avec médiane")
```

### c) Nettoyer les variables qualitatives avec mode:

```python
data[cat_cols] = data[cat_cols].apply(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 'Unknown'))
print("✓ Variables qualitatives nettoyées avec mode")
```

### d)vérifiaction finale:

```python
# VÉRIFICATION FINALE
print("\nVÉRIFICATION APRÈS NETTOYAGE:")
missing_after = data.isnull().sum()
total_missing = missing_after.sum()

print(f"Valeurs manquantes totales: {total_missing}")

if total_missing == 0:
    print("---->DATASET COMPLÈTEMENT NETTOYÉ!")
else:
    print("Colonnes avec valeurs manquantes restantes:")
    print(missing_after[missing_after > 0])
```

```python
print("\n=== STATISTIQUES FINALES ===")
print(f"Dimensions: {data.shape}")
print(f"Colonnes quantitatives: {len(data.select_dtypes(include=['int64', 'float64']).columns)}")
print(f"Colonnes qualitatives: {len(data.select_dtypes(include=['object', 'bool']).columns)}")
print(f"Valeurs manquantes: {data.isnull().sum().sum()}")
```

### Conversion des colonnes numériques contenant du texte

```python
cols_to_convert = ['Rating', 'Price', 'Duration', 'Number of viewers']

for col in cols_to_convert:
    if col in data.columns:
        data[col] = (
            data[col]
            .astype(str)
            .str.replace(r'[^0-9.,]', '', regex=True)  # remove text
            .str.replace(',', '', regex=False)
            .replace('', np.nan)
        )
        data[col] = pd.to_numeric(data[col], errors='coerce')
        data[col] = data[col].fillna(data[col].median())
print("✅ Converted numeric text fields and filled missing with median")        
```

### Extraction des attributs clés

```python
keep_cols = [
    'Title', 'Category', 'Language', 'Site', 'Level',
    'Price', 'Rating', 'Duration', 'Skills',
    'Number of viewers', 'Instructors', 'Program Type',
    'Is_Free', 'Price_Range'
]

# Keep only columns that exist in your dataset
keep_cols = [col for col in keep_cols if col in data.columns]

data_selected = data[keep_cols].copy()

print("✅ Final dataset prepared with columns:")
print(data_selected.columns.tolist())
```

### Vérification finale

```python
print("\n--- MISSING VALUES AFTER CLEANING ---")
print(data_selected.isnull().sum())
print("\nShape:", data_selected.shape)
display(data_selected.head(10))
```
